```python
workouts.groupby("EXERCISE_NAME").agg({
    "REPS": ["mean", "max", "min"],
    "WEIGHT_LBS": ["mean", "max", "min"],
    "RPE": ["mean", "max", "min"]
})
```

```python
workouts.columns
```

```python
workouts_daily = workouts.set_index("DATE").resample("D").agg({
    "DURATION_MIN": "max",
    "REPS": "sum",
    "WEIGHT_LBS": "sum",
    "RPE": "mean",
    "VOLUME": "sum"
})
workouts_daily["EXERCISED"] = (workouts_daily["REPS"] > 0).astype(int)
```
