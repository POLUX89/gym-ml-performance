```python
workouts_daily = workouts.set_index("DATE").resample("D").agg({
    "DURATION_MIN": "max",
    "REPS": "sum",
    "WEIGHT_LBS": "sum",
    "RPE": "mean",
    "VOLUME": "sum"
})
workouts_daily["EXERCISED"] = (workouts_daily["REPS"] > 0).astype(int)
```

```python
squat_heavy = workouts.loc[(workouts["EXERCISE_NAME"] == "Squat (Barbell)") &
             (workouts["WEIGHT_LBS"] > 135), ["DATE", "WEIGHT_LBS", "REPS", "RPE"]]
```

```python
squat_heavy.dtypes
```

```python
def lbs_to_kg(lbs):
    return lbs * 0.453592

squat_heavy["WEIGHT_KG"] = squat_heavy["WEIGHT_LBS"].apply(lbs_to_kg)
```
